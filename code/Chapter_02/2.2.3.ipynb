```python
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import numpy as np

#采样点数
N = 100
# 采样间隔2 ms
dt =.002

t = np.linspace(0.0,(N-1)*dt, N)
noise = np.random.randn(100)*0.5
#利用10hz、40hz和80hz正弦信号构造一个带噪音信号
x = np.sin(10.0*2.0*np.pi*t+np.pi*30/180) + np.sin(40.0*2.0*np.pi*t) + \
0.5*np.sin(80.0*2.0*np.pi*t) + noise

#傅氏变换
fft_x = fft(x)
amp_x = abs(fft_x)/len(x)*2
amp = amp_x[0:int(len(x)/2)]
label_x = np.linspace(0,int(len(x)/2)-1,int(len(x)/2))
fs = 1/dt
fre = label_x/len(x)*fs
pha_x = np.unwrap(np.angle(fft_x))
pha = pha_x[0:int(len(x)/2)]
#绘制原始信号
plt.figure(figsize=(20,10))
plt.plot(t,x)
plt.ylabel('Amplitude',fontsize=14)
plt.xlabel('Time(s)',fontsize=14)
#绘制振幅谱
plt.figure(figsize=(20,10))
plt.plot(fre,amp)
plt.ylabel('Amplitude',fontsize=14)
plt.xlabel('Frequency(hz)',fontsize=14)
plt.show()
```
