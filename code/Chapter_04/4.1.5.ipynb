## 高斯朴素贝叶斯

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB as GNB
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
digits = load_digits()
x, y = digits.data, digits.target
xtrain,xtest,ytrain,ytest = train_test_split(x, y, test_size=0.3, random_state=310)
```

```python
gnb = GNB().fit(xtrain, ytrain)
test_score = gnb.score(xtest, ytest)
print(test_score)
test_pred = gnb.predict(xtest)
print(test_pred)
test_prob = gnb.predict_proba(xtest)
print(test_prob)
confusion_matrix(ytest,test_pred)
```

## TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer as TFIDF
from sklearn.naive_bayes import MultinomialNB,ComplementNB,BernoulliNB
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics import classification_report
import pandas as pd
```

```python
categories = ["sci.space", "comp.sys.ibm.pc.hardware", "rec.sport.baseball", "talk.politics.guns"]
train_data = fetch_20newsgroups(subset="train", categories=categories)
test_data = fetch_20newsgroups(subset="test", categories=categories)
print(train_data.target_names)
print('*'*80)
for i in [0,1,2,3]:
    print(len(train_data.target),(train_data.target==i).sum())
print('*'*80)
print(train_data.data[0])
```

```python
xtrain, ytrain = train_data.data, train_data.target
xtest, ytest =  test_data.data, test_data.target
```

```python
tfidf = TFIDF().fit(xtrain)
xtrain_vec = tfidf.transform(xtrain)
xtest_vec = tfidf.transform(xtest)
temp_for_see = pd.DataFrame(xtrain_vec.toarray(), columns=tfidf.get_feature_names())
temp_for_see.head()
```

```python
tfidf_stop_words = TFIDF(stop_words='english').fit(xtrain)
xtrain_stop_vec = tfidf_stop_words.transform(xtrain)
xtest_stop_vec = tfidf_stop_words.transform(xtest)
temp_for_see = pd.DataFrame(xtrain_stop_vec.toarray(), columns=tfidf_stop_words.get_feature_names())
temp_for_see.head()
```

```python
names = ["Multinomial", "Complement", "Bernoulli"]
models = [MultinomialNB(), ComplementNB(), BernoulliNB()]
for name, model in zip(names, models):
    model.fit(xtrain_vec,ytrain)
    y_pred = model.predict(xtest_vec)
    print(name)
    print(model.score(xtest_vec, ytest))
    print(classification_report(y_pred, ytest))
```

```python
for name, model in zip(names, models):
    model.fit(xtrain_stop_vec,ytrain)
    y_pred = model.predict(xtest_stop_vec)
    print(name)
    print(model.score(xtest_stop_vec, ytest))
    print(classification_report(y_pred, ytest))
```
